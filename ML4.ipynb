```python
# --- Step 1: Import Libraries ---
import numpy as np
import matplotlib.pyplot as plt

# --- Step 2: Define the function and its derivative ---
def function(x):
    return (x + 3) ** 2

def derivative(x):
    return 2 * (x + 3)

# --- Step 3: Initialize Gradient Descent parameters ---
x = 2                   # Starting point
learning_rate = 0.1     # Step size
precision = 0.0001      # Convergence criterion
max_iterations = 1000   # Maximum number of iterations

# --- Step 4: Gradient Descent Algorithm ---
for i in range(max_iterations):
    grad = derivative(x)
    x_new = x - learning_rate * grad

    # Check for convergence
    if abs(x_new - x) < precision:
        print(f"Converged at iteration {i}")
        break

    x = x_new

# --- Step 5: Print Results ---
print(f"\nLocal minimum occurs at x = {x:.4f}")
print(f"Function value at local minimum: y = {function(x):.4f}")

# --- Step 6: Visualization ---
x_vals = np.linspace(-10, 5, 100)
y_vals = function(x_vals)

plt.plot(x_vals, y_vals, label='y = (x + 3)Â²')
plt.scatter(x, function(x), color='red', label='Local Minimum')
plt.title("Gradient Descent: Finding Local Minimum")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
